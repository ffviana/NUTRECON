```python
Flavors = ['a','b','c','d']   # CHANGE HERE

Fractals = [1,2,3,4,5,6]
FlavorOrder = Flavors
shuffle(FlavorOrder)
shuffle(Fractals)

for i in range(4):
  print('Fractal {} -> Flavor {}'.format(Fractals[i], FlavorOrder[i]))

print('\n')

DTorder = FlavorOrder * 3
shuffle(DTorder)
print('Order for DT test 1:\n\t{}'.format(DTorder))
print(80*'=')
shuffle(DTorder)
print('Order for DT test 2:\n\t{}'.format(DTorder))
```
